## What You'll Learn

This interactive visualization demonstrates:

- **Master-Slave Operation**: How two D latches work in tandem
- **Edge Triggering**: Why output only changes on the clock edge
- **Timing Details**: Setup time, hold time, and propagation delay at nanosecond resolution
- **Signal Propagation**: How data flows through the circuit stages

## How to Use

1. Use the **Next** and **Previous** buttons to step through time
2. Watch how signals propagate through the master and slave latches
3. Observe the critical clock edge where data is captured
4. Notice the delay between clock edge and output change
